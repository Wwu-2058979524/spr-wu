做两张彩色图，一张灰度图

```python
#彩色1
import cv2
import numpy as np
from PIL import Image, ImageDraw, ImageFont# 引入 ImageDraw 
img0=cv2.imread(r"C:\Users\Lenovo\Downloads\1.bmp")
img = cv2.GaussianBlur(img0, (3,3),0)#高斯模糊
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(3, 3))
imgp = cv2.dilate(img,kernel) #膨胀
imggray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)#灰度
ret,thresh=cv2.threshold(imggray,176,255,0)#二值化，指定的阈值，大于阈值的像素被分配为maxval，小于阈值的像素被分配为0
contours,hierarchy=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
#findcontour()函数中有三个参数，第一个是原图像，第二个是轮廓检索模式，第三个是轮廓逼近方法，输出等高线和层次结构。
for i in range(len(contours)):
	area = cv2.contourArea(contours[i])#计算面积
	if area>14500 and area<20000:
		a=i#用面积筛选轮廓
cnt=contours[a]
epsilon = 0.001 * cv2.arcLength(cnt, True)
# 获取近似轮廓，
approx = cv2.approxPolyDP(cnt, epsilon,True)
cv2.drawContours(img,[approx],-1,(255,0,0),3)
m=cv2.moments(cnt)
cx=int(m['m10']/m['m00'])
cy=int(m['m01']/m['m00'])#计算中心
area=cv2.contourArea(cnt)#计算面积
perimeter=cv2.arcLength(cnt,True)#计算周长
if (isinstance(img, np.ndarray)):  # 判断是否 OpenCV 图片类型
    imgPIL = Image.fromarray(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))#实现array到image的转换
pos = (50, 20)  # (left, top)，字符串左上角坐标
color = (255, 255, 255)  # 字体颜色
textSize = 40#字体大小
drawPIL = ImageDraw.Draw(imgPIL)# 创建 Draw 对象
fontText = ImageFont.truetype("font/simsun.ttc", textSize, encoding="utf-8")
#创建一个字体对象，encoding-使用哪种字体编码
drawPIL.text(pos, f'吴兆芸\n轮廓面积:{area}\n周长:{perimeter}\n中心:({cx},{cy})', color, font=fontText)
# 第一个参数：指定文字区域的左上角在图片上的位置(x,y)
# 第二个参数：文字内容
# 第三个参数：颜色
# 第四个参数：字体

imgPutText = cv2.cvtColor(np.asarray(imgPIL), cv2.COLOR_RGB2BGR)
#实现image到array的转换
cv2.imshow("imgPutText", imgPutText)  # 显示叠加图像 imgAdd
key = cv2.waitKey(0)  # 等待按键命令
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#高斯滤波
def GausBlur(img):
    gaus = cv2.GaussianBlur(img,(5,5),2)   #(5, 5)表示高斯矩阵的长与宽都是5，标准差取2
    return gaus

#肤色检查处理
def skin_img(gaus_img):
    ycrcb= cv2.cvtColor(gaus_img, cv2.COLOR_BGR2YCrCb) # 把图像转换到YUV色域
    (y, cr, cb) = cv2.split(ycrcb) # 图像分割, 分别获取y, cr, br通道图像
    cr1 = cv2.GaussianBlur(cr,(3,3),0) # 对cr通道分量进行高斯滤波#高斯滤波, cr 是待滤波的源图像数据, (5,5)是值窗口大小, 0 是指根据窗口大小来计算高斯函数标准差
	# 根据OTSU算法求图像阈值, 对图像进行二值化_, 
    ret,skin = cv2.threshold(cr1, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    return skin
#开运算操作

#定义一个图像处理函数
def img_disapose():
    #捕获存储完图像之后 对图像进行处理
    #读取上面存储的图片
    img = cv2.imread('cap_RGB.jpg')   #0为灰度，1为彩色
    img = cv2.resize(img,(640,480))   #设置窗口大小
    #高斯滤波
    gaus_img = GausBlur(img)
    #肤色检测处理
    skin= skin_img(gaus_img)
    #轮廓检测
    contours, hierarchy = cv2.findContours(skin, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)  
    #描绘轮廓
    for i in range(len(contours)):
        area = cv2.contourArea(contours[i])#计算面积
        if area>500 :
            a=i#用面积筛选轮廓
    cnt=contours[a]   
    epsilon = 0.001 * cv2.arcLength(cnt, True)
    # 获取近似轮廓，
    approx = cv2.approxPolyDP(cnt, epsilon,True)
    cv2.drawContours(frame,[approx],-1,(255,0,0),3)
    area=cv2.contourArea(cnt)
    hull = cv2.convexHull( cnt )
    cv2.drawContours(frame,[hull],-1,(255,0,0),3)
    areahull=cv2.contourArea(hull)
    a=areahull/area
    if a<1.2:
	    m=0
    elif a<1.5:
        m=1
    else:
	    m=2
    text = ["rock", "scissors", "paper"]
    cv2.putText(frame, text[m], (30,30), cv2.FONT_HERSHEY_SIMPLEX, 1,(0,0,255),2)
cv2.namedWindow("Photo_Detect")  #定义一个窗口
cap=cv2.VideoCapture(0) #捕获摄像头图像  0位默认的摄像头 笔记本的自带摄像头  1为外界摄像头
while(True):                  #值为1不断读取图像
    ret, frame = cap.read()  #视频捕获帧
    cv2.imwrite('cap_RGB.jpg',frame) #写入捕获到的视频帧  命名为cap_RGB.jpg
    img_disapose()  #图像处理
    cv2.imshow('Photo_Detect',frame)     #显示窗口 查看实时图像
    if cv2.waitKey(1) & 0xFF == ord('q'):   #按Q关闭所有窗口  一次没反应的话就多按几下
        break
#执行完后释放窗口
cap.release()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
